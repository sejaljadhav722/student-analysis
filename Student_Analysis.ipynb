```python
plt.figure(figsize= (5,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title ("Gender Distribution")
plt.show()
```

```python
# from the above chart we have analysed that :
#the number of female in the data is more than the number of males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean', "WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize= (4,4))
sns.heatmap(gb, annot = True)
plt.title ("Relationship Between Parents Education And Student's Score")
plt.show()
```

```python
#from the above chart we have concluded that the education of the parents have a good impact on the Student's Score
```
